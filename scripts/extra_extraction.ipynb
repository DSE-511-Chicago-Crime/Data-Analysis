```python
import bq_helper
from bq_helper import BigQueryHelper
# https://www.kaggle.com/sohier/introduction-to-the-bq-helper-package
chicago_crime = bq_helper.BigQueryHelper(active_project="bigquery-public-data",
                                   dataset_name="chicago_crime")
```

```python
# connect to the crime database
bq_assistant = BigQueryHelper("bigquery-public-data", "chicago_crime")
bq_assistant.list_tables()
```

```python
for year in ['2015', '2016', '2017', '2018']:
    query = """
            SELECT unique_key, date, block, 
            iucr, primary_type, description,
            location_description, arrest, domestic,
            fbi_code, year, district
            FROM `bigquery-public-data.chicago_crime.crime`
            WHERE year = 
            """ + year
    response = chicago_crime.query_to_pandas_safe(query)
    response.to_csv(year + '_crime.csv')
```
